```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
from datetime import date
today = date.today()
date = today.strftime("%Y%m%d/")

#def Scrape(a):
page  = 'https://coinmarketcap.com/historical/'+ date
page
page = requests.get(page).text
soup = BeautifulSoup(page, 'html.parser')
tr = soup.find_all('tr', attrs={'class':'cmc-table-row'})

#defining the lists for storing each column
crypto_Name = []
Market_Cap = []
Crypto_Price = []
Su = []
Sy = []
Volume = []


number = 20
count = 0
for row in tr:
    if count == number:
        break;
    count = count + 1

# Scraping all the values            
    name = row.find('td', attrs={'class':'cmc-table__cell cmc-table__cell--sticky cmc-table__cell--sortable cmc-table__cell--left cmc-table__cell--sort-by__name'})
    crypto_name = name.find('a', attrs={'class':'cmc-table__column-name--name cmc-link'}).text.strip()
    market_cap = row.find('td', attrs={'class':'cmc-table__cell cmc-table__cell--sortable cmc-table__cell--right cmc-table__cell--sort-by__market-cap'}).text.strip()
    crypto_price = row.find('td', attrs={'class':'cmc-table__cell cmc-table__cell--sortable cmc-table__cell--right cmc-table__cell--sort-by__price'}).text.strip()

    circulatingSupplySymbol = row.find('td', attrs={'class':'cmc-table__cell cmc-table__cell--sortable cmc-table__cell--right cmc-table__cell--sort-by__circulating-supply'}).text.strip()
    supply = circulatingSupplySymbol.split(' ')[0]
    sym = circulatingSupplySymbol.split(' ')[1]

    vol = row.find('td', attrs = {'class' : "cmc-table__cell cmc-table__cell--sortable cmc-table__cell--right cmc-table__cell--sort-by__volume-24-h"}).text.strip()

#appeding to the list
    crypto_Name.append(crypto_name)
    Market_Cap.append(market_cap)
    Crypto_Price.append(crypto_price)
    Su.append(supply)
    Sy.append(sym)
    Volume.append(vol)
```

```python
for i in [crypto_Name, Market_Cap, Crypto_Price, Su, Sy, Volume]:
    print(i,'\n')
```

```python
df = pd.DataFrame()
```

```python
df['Name'] = crypto_Name
df['Market Cap'] = Market_Cap
df['Price'] = Crypto_Price
df['Circulating Suppy'] = Su
df['Symbol'] = Sy
df['Volume'] = Volume
df
```

```python
df.to_csv('crypto.csv')
```
